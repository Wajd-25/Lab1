```python
# Example 5.1

from google.colab import files
uploaded = files.upload()

from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np

image = imread('Wajd.jpg')

print("Shape:", image.shape)
print("Data type:", image.dtype)

plt.imshow(image)
plt.title("Original Image")
plt.axis('off')
plt.show()

if len(image.shape) == 3:
    gray = image.mean(axis=2)
else:
    gray = image

plt.imshow(gray, cmap='gray')
plt.title("Grey Scale Image")
plt.axis('off')
plt.show()
```

```python
# Example 6.1

min_val = gray.min()
max_val = gray.max()

enhanced = (gray - min_val) / (max_val - min_val) * 255

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Original")

plt.subplot(1,2,2)
plt.imshow(enhanced, cmap='gray')
plt.title("Contrast Enhanced")

plt.show()
```

```python
# Example 7.1

noise = np.random.normal(0, 20, gray.shape)
noisy = gray + noise

plt.imshow(noisy, cmap='gray')
plt.title("Noisy Image")
plt.show()
```

```python
# Example 7.2

def mean_filter(img, k=3):
    pad = k // 2
    padded = np.pad(img, pad, mode='edge')
    output = np.zeros_like(img)

    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            output[i, j] = padded[i:i+k, j:j+k].mean()

    return output

denoised = mean_filter(noisy)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(noisy, cmap='gray')
plt.title("Noisy")

plt.subplot(1,2,2)
plt.imshow(denoised, cmap='gray')
plt.title("Denoised")

plt.show()
```

```python
# Example 8.1

threshold = gray.mean()
binary = gray > threshold

plt.imshow(binary, cmap='gray')
plt.title("Binary Segmentation")
plt.show()
```

```python
# Exercise 2.1  Visualising an Image Histogram
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from google.colab import files


uploaded = files.upload()
image_name = list(uploaded.keys())[0]

image = io.imread(image_name, as_gray=True)
image = (image * 255).astype(np.uint8)


plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title("Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.hist(image.flatten(), bins=256, range=(0,255))
plt.title("Histogram")
plt.xlabel("Grey level")
plt.ylabel("Frequency")

plt.show()
```

```python
# Exercise 3.1  Computing a Normalised Histogram

hist, bins = np.histogram(image.flatten(), bins=256, range=(0,255))
hist_norm = hist / hist.sum()

print(hist_norm.sum())
```

```python
# Exercise 4.1  Manual Histogram Equalisation

def hist_equalise(img):
    hist, _ = np.histogram(img.flatten(), bins=256, range=(0,255))
    cdf = hist.cumsum()

    cdf_norm = (cdf - cdf.min()) * 255 / (cdf.max() - cdf.min())
    cdf_norm = cdf_norm.astype(np.uint8)

    return cdf_norm[img]

equalised = hist_equalise(image)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(equalised, cmap='gray')
plt.title("Equalised Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.hist(equalised.flatten(), bins=256, range=(0,255))
plt.title("Equalised Histogram")

plt.show()
```

```python
# Exercise 6.1  Image Negative

negative = 255 - image

plt.imshow(negative, cmap='gray')
plt.title("Negative Image")
plt.axis('off')
plt.show()
```

```python
# Exercise 7.1  Binary Thresholding

threshold = 128

binary = np.zeros_like(image)
binary[image > threshold] = 255

plt.imshow(binary, cmap='gray')
plt.title("Thresholded Image")
plt.axis('off')
plt.show()
```

```python
# Exercise 8.1  Log Transformation
image_norm = image / 255.0

c = 1 / np.log(1 + np.max(image_norm))
log_img = c * np.log(1 + image_norm)

plt.imshow(log_img, cmap='gray')
plt.title("Log Transformed Image")
plt.axis('off')
plt.show()
```

```python
# Exercise 8.2  Gamma (Power Law) Transformation

gamma = 0.4
gamma_img = np.power(image_norm, gamma)

plt.imshow(gamma_img, cmap='gray')
plt.title("Gamma Correction ( = 0.4)")
plt.axis('off')
plt.show()
```

```python
# Exercise 9.1  Mean (Averaging) Filter

from scipy.ndimage import uniform_filter

smoothed = uniform_filter(image, size=3)

plt.imshow(smoothed, cmap='gray')
plt.title("Averaging Filter (33)")
plt.axis('off')
plt.show()
```

```python
# Exercise 9.2  Median Filter

from scipy.ndimage import median_filter

median = median_filter(image, size=3)

plt.imshow(median, cmap='gray')
plt.title("Median Filter")
plt.axis('off')
plt.show()
```

```python
# Exercise 10.1  Laplacian Sharpening

from scipy.ndimage import laplace

lap = laplace(image)

sharpened = image - lap

sharpened = np.clip(sharpened, 0, 255)
sharpened = sharpened.astype(np.uint8)

plt.imshow(sharpened, cmap='gray')
plt.title("Laplacian Sharpening")
plt.axis('off')
plt.show()
```

```python
# Exercise 10.2  Sobel Edge Detection

from scipy.ndimage import sobel

sx = sobel(image, axis=0)
sy = sobel(image, axis=1)

edges = np.hypot(sx, sy)

plt.imshow(edges, cmap='gray')
plt.title("Sobel Edge Magnitude")
plt.axis('off')
plt.show()
```

```python
# Mini Project  Multi Step Image Enhancement

# 1) Import Libraries-

import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from scipy.ndimage import uniform_filter, laplace
from google.colab import files

# 2) Upload and Read Image

uploaded = files.upload()
image_name = list(uploaded.keys())[0]

image = io.imread(image_name, as_gray=True)
image = (image * 255).astype(np.uint8)

# 3) Histogram Equalisation

def hist_equalise(img):
    hist, _ = np.histogram(img.flatten(), bins=256, range=(0,255))
    cdf = hist.cumsum()
    cdf_norm = (cdf - cdf.min()) * 255 / (cdf.max() - cdf.min())
    cdf_norm = cdf_norm.astype(np.uint8)
    return cdf_norm[img]

equalised = hist_equalise(image)

# 4) Smoothing (Mean Filter)

smoothed = uniform_filter(equalised, size=3)

# 5) Sharpening (Laplacian)

lap = laplace(smoothed)
sharpened = smoothed - lap

sharpened = np.clip(sharpened, 0, 255)
sharpened = sharpened.astype(np.uint8)

# 6) Display Results

plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(equalised, cmap='gray')
plt.title("Histogram Equalised")
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(smoothed, cmap='gray')
plt.title("After Smoothing")
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(sharpened, cmap='gray')
plt.title("Final Enhanced Image")
plt.axis('off')

plt.tight_layout()
plt.show()
```
